# 데이터 전처리, [분석 이벤트 생성] 세션 데이터 생성

```python
#### user_id, event_at 정렬 후
#### event_at 추출
np_df_raw_event_at = pd.to_datetime(df_raw_____['event_at']).to_numpy()
```

```python
#### event_at 한 칸 내림
np_df_raw_event_at_dbo = np.append(
    pd.to_datetime(df_raw_____['event_at'][:1]).to_numpy()
    ,pd.to_datetime(df_raw_____['event_at'].values[:-1]).to_numpy()
)
```

```python
#### event_at - event_at (한 칸 내림))
np_diff_raw_event_at = (np_df_raw_event_at - np_df_raw_event_at_dbo).astype('timedelta64[s]').astype(int)
```

```python
#### user_id 추출
list_of_car_id_in_df_raw = df_raw_____['user_id'].values.astype(int)
```

```python
result_of_df_raw_sessions = []
first_car_id_in_df_raw = list_of_car_id_in_df_raw[0]
num = 1
```

```python
#### user_id 별 이벤트 발생 시간 순차적 정렬 후
#### 동일한 user_id, 직전 이벤트와의 시간 차이가 10분 (600s) 이하의 경우 동일 세션으로 정의
#### 동일 세션일 경우 같은 숫자를 부여, 
#### 10분 이상이 될 경우 기존 숫자 + 1

for x, y in zip(list_of_car_id_in_df_raw, np_diff_raw_event_at):
    if x == first_car_id_in_df_raw:
        if y <= 600:
            result_of_df_raw_sessions.append(num)
            first_car_id_in_df_raw = x
        else:
            num = num + 1
            first_car_id_in_df_raw = x
            result_of_df_raw_sessions.append(num)
    else:
        first_car_id_in_df_raw = x
        num = 1
        result_of_df_raw_sessions.append(num)
```

```python
df_raw_____['sessions'] = result_of_df_raw_sessions
```

# 데이터 전처리, [분석 이벤트 생성] 가입 후 {이벤트} 발생 경과일 생성

```python
temp_list = [
    'car_create'
    ,'price_s****'
    ,'mil****'
    ,'trans****'
    ,'opt****'
    ,'loca****'
    ,'acci****'
    ,'cond****'
    ,'share_p****'
    ,'similar_c****'
    ,'revi****'
]

dict_test = {}
for event in temp_list:
    print(df_raw_____[
        (df_raw_____['event_type']==event)
    ].shape)

    print(df_raw_____[
        (df_raw_____['event_type']==event)
    ].drop_duplicates(['car_id']).shape)

    temp_car_id = df_raw_____[
        (df_raw_____['event_type']==event)
    ].sort_values('event_at',ascending=False).drop_duplicates('car_id')['car_id'].astype(int).values

    temp_np_sign_at = pd.to_datetime(
        df_raw_____[
            (df_raw_____['event_type']==event)
        ].sort_values('event_at',ascending=False).drop_duplicates('car_id')['sign_up_at']
    ).to_numpy()

    temp_np_event_at = pd.to_datetime(
        df_raw_____[
            (df_raw_____['event_type']==event)
        ].sort_values('event_at',ascending=False).drop_duplicates('car_id')['event_at']
    ).to_numpy()

    temp_np_diff = ((temp_np_event_at - temp_np_sign_at).astype('timedelta64[s]').astype(int) / 86400).astype(int)

    dict_test[event] = pd.DataFrame(index=temp_car_id,data=temp_np_diff).reset_index().rename(columns={'index':'car_id',0:'d_s2%s'%event})
    print(dict_test[event].shape)
```

```python
for key in list(dict_test.keys()):
    print(df_raw_____.shape)
    df_raw_____  = pd.merge(
        df_raw_____
        ,dict_test[key]
        ,'left'
        ,on='car_id'
        ,suffixes=('','_y')
    )
    print(df_raw_____.shape)    
```
