```python
df = pd.concat([
    pd.read_csv('/path/0_인앱이벤트.csv')
    ,pd.read_csv('/path/1_인앱이벤트.csv')
])
```

```python
sql = """
SELECT
`Advertising ID`
,`Customer User ID`
,`Media Source`
,`Campaign`
,`Site ID`
,`Campaign ID`
,`Channel`
,`Adset`
,`Adset ID`
,`Ad`
,`Ad ID`
,`Match Type`
,`Install Time`
,`Platform`

FROM ****
"""

cursor.execute(sql)
columns = [description[0] for description in cursor.description]
df_non_organic_installs = pd.DataFrame(cursor.fetchall(),columns=columns)

sql = """
SELECT
`Advertising ID`
,`Customer User ID`
,`Media Source`
,`Campaign`
,`Site ID`
,`Campaign ID`
,`Channel`
,`Adset`
,`Adset ID`
,`Ad`
,`Ad ID`
,`Install Time`
,`Platform`

FROM ****
"""

cursor.execute(sql)
columns = [description[0] for description in cursor.description]
df_organic_installs = pd.DataFrame(cursor.fetchall(),columns=columns)
```
